#5.Filtering

Filtering means selecting rows based on conditions.

```python
df[df["Salary"] > 20000]
```

```python
df[df['Department'] == "IT"]
```

```python
#and
df[(df['Salary'] < 30000) & (df['Department'] == "IT")]
```

```python
# or
df[(df['ID'] > 104) | (df['Department'] == "HR")]
```

```python
#not equal
df[df["Department"] != "HR"]
```

```python
#isin
df[df['Department'].isin(["IT" , "HR"])]
```

```python
#between()
df[df['Salary'].between(20000,50000)]
```

#12. fillna()
this method is used to replace missing (NaN) values.


```python
#replace with fixed value
df.fillna(0)
```

```python
#Fill with mean
df["Age"].fillna(df["Age"].mean())
```

```python
#Backward fill(bfill)
df.fillna(method="bfill")
```

#13.dropna()
The dropna() method removes rows or columns containing missing values.

```python
df
```

```python
#remove rows by default
df.dropna()
```

```python
#remove columns
df.dropna(axis=1)
```

The how parameter tells Pandas when to remove them.

how="any" (default) → Remove if at least one value is NaN.
how="all" → Remove only if every value in the row (or column) is NaN.

#22.concat():
concat() is used to combine (stack) two or more DataFrames.

There are two ways to concatenate:

Vertically (one below another) → axis=0 (default)

Horizontally (side by side) → axis=1

```python
#vertically
pd.concat([df,location])
```

```python
#horizontally
pd.concat([df,location],axis=1)
```
