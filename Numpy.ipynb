##10.Indexing

###1.positive Indexing

```python
arr = np.array([10,20,30,40,50])

print("First Element:- ",arr[0])
print("Second Element:- ",arr[1])
print("Third Element:- ",arr[2])
print("Fourth Element:- ",arr[3])
print("Fourth Element:- ",arr[4])
```

###2.Negative Indexing

```python
print("First Element:- ",arr[-5])
print("Second Element:- ",arr[-4])
print("Third Element:- ",arr[-3])
print("Fourth Element:- ",arr[-2])
print("Fourth Element:- ",arr[-1])
```

##11.Slicing

###1.Positive Slicing

```python
print(arr[0:5])
print(arr[:3])
print(arr[2:])
print(arr[1:4])
```

###2.Negative Slicing

```python
print(arr[-5:])
print(arr[:-2])
print(arr[-3:])
print(arr[-4:-1])
```
