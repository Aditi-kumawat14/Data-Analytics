```python
import matplotlib.pyplot as plt
```

```python
months = ["jan","feb","march","april","may"]
sales = [10000,50000,30000,20000,40000]
```

#1.Line chart

```python
plt.plot(months,sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
#add markers
plt.plot(months,sales , marker="*")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
#Add a Legend
#add markers
plt.plot(months,sales , marker="*" , label="Sales" , color="red")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#Add a line style
plt.plot(months,sales , linestyle="--")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python

#Add a line width
plt.plot(months,sales , linewidth=6)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#Add a marker size , marker face color , marker edge color
plt.plot(months,sales , marker="*" , markersize=12 , markerfacecolor="red" , markeredgecolor="black")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
#Add a grid
plt.plot(months,sales )
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
#Add a grid color
plt.plot(months,sales )
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(color="yellow")
plt.show()
```

```python
#Add a figure
plt.plot(months,sales )
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.figure(figsize=(20,10))
plt.show()
```

```python
#save graph
plt.plot(months,sales )
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.savefig("monthly_sales.png")
plt.show()
```

#2.Bar Chart

```python
students = ["A","B","C","D","E"]
marks = [50,60,70,80,90]

plt.bar(students,marks)
plt.show()
```

```python
#add title , xlabel , ylabel
plt.bar(students,marks)
plt.title("Students Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#add color , width
plt.bar(students,marks , color="gray" , width=0.5)
plt.title("Students Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#add edge color , line width
plt.bar(students,marks ,color="gray", edgecolor="black" , linewidth=4)
plt.title("Students Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#add transparency
plt.bar(students,marks , color="gray" , alpha = 0.5 , edgecolor="black")
plt.title("Students Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#add grid
plt.bar(students,marks , color="gray" , alpha = 0.5 , edgecolor="black")
plt.title("Students Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.show()
```

```python
#add grid
plt.figure(figsize=(8,5))

plt.bar(
    students,
    marks,
    color="green",
    edgecolor="black",
    linewidth=2,
    width=0.5,
    alpha=0.8,
    label="Marks"
)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.legend()
plt.show()
```

#4.Scatter Plot

```python
hrs = [1,2,3,4,5]
marks = [40,50,60,75,90]
```

```python
plt.scatter(hrs,marks , color="red")
plt.title("Hours of study VS marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#add marker size , marker
plt.scatter(hrs,marks , marker="*", s =300)
plt.title("Hours of study VS marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#add color and border
plt.scatter(hrs,marks ,s = 300 , color="red" ,edgecolors="black" ,linewidth=5)
plt.title("Hours of study VS marks")
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()
```

```python
#add c and cmap
plt.scatter(hrs,marks ,s = 300 , c=[1,2,3,4,5] ,cmap="rainbow" )
plt.title("Hours of study VS marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#add color bar
plt.scatter(hrs,marks ,s = 300 , c=[1,2,3,4,5] ,cmap="rainbow" )
plt.title("Hours of study VS marks")
plt.colorbar()
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()
```

```python
x = [1,2,3]
y = [40,60,80]

plt.scatter(x, y)

plt.annotate("A", (1,40))
plt.annotate("B", (2,60))
plt.annotate("C", (3,80))

plt.show()
```

#4.Histogram

```python
marks = [45,50,55,60,65,70,70,75,80,85,90]
```

```python
plt.hist(marks)
plt.title("Students Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

#5.Pie Chart

```python
hours = [8,7,5,2,3]
activities = ["sleep","Study","Games", "Exercise", "Others"]
```

```python
plt.pie(hours)
plt.show()
```

```python
#add labels
plt.pie(hours , labels=activities)
plt.show()
```

```python
#add percentages
plt.pie(hours , autopct="% 1.2f %%")
plt.show()
```

```python
#add colors
clr = ["red","blue","yellow","gray","green"]
plt.pie(hours , colors= clr)
plt.show()
```

```python
#add Explode
exp = [0,0.2,0,0.3,0]
plt.pie(hours , explode=exp)
plt.show()
```

```python
# add shadow
plt.pie(hours , shadow=True)
plt.show()
```

```python
# start angle
plt.pie(hours , startangle=90)
plt.show()
```

```python
# add title
plt.pie(hours , labels=activities)
plt.title("Activity hours")
plt.show()
```

#6.Box plot

```python
marks = [40,45,50,55,60,65,70,75,80]
```

```python
plt.boxplot(marks)
plt.title("Students Marks")
plt.ylabel("Marks")
plt.show()
```

```python
#add color = A box plot doesn't have a simple color= parameter.
plt.boxplot(marks , patch_artist=True)
plt.title("Students Marks")
plt.ylabel("Marks")
plt.show()
```

```python
#show mean
plt.boxplot(marks , showmeans=True)
plt.title("Students Marks")
plt.ylabel("Marks")
plt.show()
```

```python
#hide outlies bez by default it is shown
plt.boxplot(marks , showfliers=False)
plt.title("Students Marks")
plt.ylabel("Marks")
plt.show()
```

```python
#Horizontal Box Plot
plt.boxplot(marks , vert=False)
plt.title("Students Marks")
plt.ylabel("Marks")
plt.show()
```
