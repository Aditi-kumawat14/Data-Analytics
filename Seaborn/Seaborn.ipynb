#Seaborn

Why do we use Seaborn?

Suppose we create a line chart in Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x,y)
plt.show()
```

Now the same graph using Seaborn style

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

plt.plot(x,y)

plt.show()
```

#1 .Line Plot

```python
#import lib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
tips = sns.load_dataset("tips")
tips
```

```python
sns.lineplot(data=tips , x = "size",y = "total_bill")
plt.show()
```

```python
#How to Remove the Blue Shaded Area?
sns.lineplot(data=tips , x = "size",y = "total_bill" , errorbar=None)
plt.show()
```

```python
#add hue
sns.lineplot(data=tips , x = "size",y = "total_bill" , hue="sex" , errorbar=None)
plt.show()
```

```python
#to hide legend and add markers , linewidth
sns.lineplot(data=tips , x = "size",y = "total_bill" , hue="sex", legend=False , errorbar=None , marker="*" , linewidth=5)
plt.show()
```

```python
#add palettes
sns.lineplot(data=tips , x = "size",y = "total_bill" , hue="sex" ,  palette="rainbow")
plt.show()
```

#2.Seaborn Scatter Plot

```python
sns.scatterplot(data=tips , x = "size",y = "total_bill")
plt.show()
```

```python
sns.scatterplot(data=tips , x = "size",y = "total_bill" , hue="sex" , style="smoker" )
plt.show()
```

#3.Bar plot

```python
sns.barplot(data=tips, x="day", y="total_bill")

plt.title("Average Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")

plt.show()
```

#4.Histogram

```python
sns.histplot(data=tips, x="total_bill")

plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

```python
#add bins
sns.histplot(data=tips, x="total_bill" ,  bins=20)

plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

```python
#add kde
sns.histplot(data=tips, x="total_bill" , kde=True)

plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

```python
#multiple :- Display multiple groups in different ways.
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    multiple="stack"
)
plt.show()
```

#5.BoxPlot

```python
sns.boxplot(data=tips , x="day" , y="total_bill" )
plt.show()
```

```python
sns.boxplot(data=tips , x="day" , y="total_bill" , hue="sex")
plt.show()
```

#6.Count Plot

```python
sns.countplot(data=tips, x="day")
plt.show()
```

```python
#add Order
sns.countplot(data=tips, x="day" , order=["Sat","Fri","Sun","Thur"])
plt.show()
```

#7. Violin Plot

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
#add split
sns.violinplot(data=tips, x="day", y="total_bill" , split=True  )
plt.show()
```

```python
#add split with hue
sns.violinplot(data=tips, x="day", y="total_bill",hue="sex" , split=True  )
plt.show()
```

```python
#add inner
sns.violinplot(data=tips, x="day", y="total_bill", inner="point" )
plt.show()
```

#8.Heatmap

```python
corr = tips.corr(numeric_only=True)
```

```python
corr
```

```python
sns.heatmap(corr , annot=True)
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=1,
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```
